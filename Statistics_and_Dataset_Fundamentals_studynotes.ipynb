## Module 1 - Introduction to Data
### What is Data?

Data is a collection of facts, numbers, words, or measurements.

Example: Student marks, age, height, salary.

### Record (Observation)

One row in a dataset is called a record.

Example: Ana, 20, 85

### Feature (Attribute)

A feature is a column in a dataset.

Examples: Name, Age, Marks.

### Variable

A variable is a value that can change.

Example: Age can be 18, 20, 25, etc.

### Label / Target Variable

The value we want to predict.

Example: Predict whether a customer will buy a product.

### Metadata

Metadata means data about data.

Example: File name, creation date, number of rows.

## Module 2 - Types of Data
### Qualitative Data

Describes qualities or categories.

Examples: Color, Gender, City.

#### Nominal Data

Categories with no order.

Example: Red, Blue, Green.

#### Ordinal Data

Categories with order.

Example: Low, Medium, High.

### Quantitative Data

Numerical data.

#### Discrete Data

Countable values.

Example: Number of students.

#### Continuous Data

Measured values.

Example: Height, Weight.

## Module 7 - Measures of Central Tendency
### Mean

Average value.

Formula: Sum of values / Number of values

Example: (10 + 20 + 30) / 3 = 20

Use: Average marks, salary.

### Median

Middle value after sorting.

Example: 10, 20, 30 → Median = 20

Use: Better when outliers exist.

### Mode

Most frequent value.

Example: 10, 20, 20, 30 → Mode = 20

Use: Popular product, common category.

## Module 9 - Distribution Basics
### Normal Distribution

Bell-shaped and balanced.

### Skewed Distribution

Data is tilted to one side.

### Symmetric Distribution

Left and right sides look similar.

### Outliers

Values that are very different from others.

Example: Most salaries are ₹30,000-₹50,000, but one salary is ₹10,00,000.

### Why important in AI/ML?

Outliers can affect model accuracy.

Many algorithms work better with normally distributed data.
